Import Libraries

```python
import numpy as np
import pandas as pd
```

Read Data

```python
df = pd.read_excel("/content/FlavoneData.xlsx")
df
```

Second and Third Columns of Table

```python
df[['authors', 'flavone', 'cell line']].groupby(['authors', 'flavone', 'cell line']).size()
```

Fourth Column of Table

```python
df[['authors', 'dose']].groupby(['authors']).min()
```

```python
df[['authors', 'dose']].groupby(['authors']).max()
```

Fifth Column of Table

```python
df[['authors', 'time']].groupby(['authors','time']).size()
```

Sixth Column of Table

```python
df[['authors', 'viability-mean', 'viability-SD']].groupby(['authors']).mean()
```

Last Column of Table

```python
df[['authors']].groupby(['authors']).size()
```
